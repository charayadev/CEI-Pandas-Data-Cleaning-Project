## 5. Missing Value Analysis and Treatment

```python
# There are Multiple columns that must be deal with the values let us fix them one by one 
# High Missing Values 
# columns such as videos customer said with the no ____ data 
df['videos'] = df['videos'].fillna('No Video')
df['what_customers_said'] = df['what_customers_said'].fillna('No Reviews')
df['variations'] = df['variations'].fillna('No Variations')
# used with the inplace but Give a Warning
```

```python
# Medium Missing 
df['seller_name'] = df['seller_name'].fillna('Unknown Seller')
df['seller_information'] = df['seller_information'].fillna('Not Provided')
```

```python
# Numercial Columns Discount know lets have the statiscial approach to fill the values
discount_data = df['discount'].dropna()
plt.boxplot(discount_data)
plt.title('Box Plot of Discount')
plt.ylabel('Discount')
plt.show()
# Display summary statistics
print("\nDiscount Column Statistics:")
print(f"Mean: {discount_data.mean():.2f}%")
print(f"Median: {discount_data.median():.2f}%")
print(f"Mode: {discount_data.mode().values[0]:.2f}%")
print(f"Standard Deviation: {discount_data.std():.2f}%")
```

**What I noticed from the boxplot:**

- The median discount is around 56% (the orange line inside the box)
- There are some outliers below 10% - these are products with unusually low discounts
- The data is slightly skewed, meaning it's not perfectly centered

**How I'll fill the missing values:**

I'm going with **median** instead of mean because:
- Mean ≈ 54% (gets pulled down by those low outlier values)
- Median ≈ 56% (better represents what most products actually get)
- Median isn't affected by extreme values, so it's safer to use here

Using median makes sure the filled values match the typical discount pattern in the dataset.

```python
# Fill missing discount values with median
df['discount'] = df['discount'].fillna(df['discount'].median())
```
